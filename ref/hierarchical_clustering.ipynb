```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris
import pandas as pd


iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

data
```

```python
# Perform hierarchical clustering with different linkage methods
linkage_single = linkage(data, method='single')
linkage_complete = linkage(data, method='complete')
linkage_average = linkage(data, method='average')

# Plot dendrograms
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
dendrogram(linkage_single)
plt.title('Single Linkage')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

plt.subplot(1, 3, 2)
dendrogram(linkage_complete)
plt.title('Complete Linkage')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

plt.subplot(1, 3, 3)
dendrogram(linkage_average)
plt.title('Average Linkage')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```
