```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(42)  # For reproducibility
a_true = 5
num_points = 100
x = np.random.uniform(0, 10, num_points)
noise = np.random.normal(0, 1, num_points)
y = a_true + noise

# Estimate 'a' by averaging y values
a_avg = np.mean(y)

# Estimate 'a' using scikit-learn LinearRegression
x_reshaped = x.reshape(-1, 1)  # Reshape x to 2D array for sklearn
model = LinearRegression(fit_intercept=True)
model.fit(np.zeros_like(x_reshaped), y)  # Fit x=0 to y (effectively y=a)
a_sklearn = model.intercept_

# Print results
print(f"True value of a: {a_true}")
print(f"Estimated a (average method): {a_avg}")
print(f"Estimated a (scikit-learn): {a_sklearn}")

# Plot the data and the fitted lines
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data points', alpha=0.6)
plt.axhline(y=a_avg, color='green', linestyle='--', label=f'Average line (a={a_avg:.2f})')
plt.axhline(y=a_sklearn, color='red', linestyle='-', label=f'Scikit-learn line (a={a_sklearn:.2f})')
plt.axhline(y=a_true, color='black', linestyle=':', label=f'True value (a={a_true})')

plt.title('Linear Regression Fit: y = a')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
